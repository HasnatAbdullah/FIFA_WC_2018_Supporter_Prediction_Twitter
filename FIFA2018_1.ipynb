```python
# Checking the missing values

print(df.isnull().sum())
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(400,300))
sns.displot(data=df.isnull().melt(value_name="missing"),hue="missing",y="variable",multiple="fill",aspect=1.25)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
plt.figure(figsize=(30,30))
df2=df.drop(["Team"],axis=1)
corr=df2.corr()
sns.heatmap(data=corr,annot=True)
```

```python
mean=df["WC"].mean()
print("Mean is {0:.2f}".format(mean))
median=df["WC"].median()
print("Median is {0:.2f}".format(median))
mode=df["WC"].mode()
print("Mode is",mode)
variance=df["WC"].var()
print("Variance is {0:.2f}".format(variance))
standard_deviation=df["WC"].std()
print("Standard Deviation is {0:.2f}".format(standard_deviation))
minimum=df["WC"].min()
print("Minimum is {0:.2f}".format(minimum))
maximum=df["WC"].max()
print("Maximum is {0:.2f}".format(maximum))
skewness=df["WC"].skew()
print("Skewness is {0:.2f}".format(skewness))
kurtosis=df["WC"].kurtosis()
print("Kurtosis is {0:.2f}".format(kurtosis))
```

```python
# Checking the Distribution of Word Counts variable


plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
f, axes = plt.subplots(1, 2)
sns.histplot(data=df["WC"],ax=axes[0])
sns.kdeplot(data=df["WC"],ax=axes[1],shade=True)
```
